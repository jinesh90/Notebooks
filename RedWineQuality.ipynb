```python
import pandas as pd
wine_db = pd.read_csv("wine_quality.csv")
wine_db.head(10)
```

```python
wine_db.columns
```

```python
wine_db = wine_db.drop(["Unnamed: 0"],axis=1)
```

```python
wine_db
```

```python
from sklearn.model_selection import train_test_split
train_set,test_set = train_test_split(wine_db,test_size=0.1,random_state=42)
train_set
```

```python
wine_test = train_set.copy()
wine_quality = wine_test["quality"].copy()
wine_test = wine_test.drop(["quality"],axis=1)
```

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor()
forest_reg.fit(wine_test,wine_quality)
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np
forest_score = cross_val_score(forest_reg,wine_test,wine_quality,scoring="neg_mean_squared_error",cv=1000)
forest_score_rmse = np.sqrt(-forest_score)
forest_score_rmse.mean()
#forest_score_rmse.std()
```
