```python
import seaborn as sns
sns.set(color_codes=True)
%matplotlib inline
```

```python
auto = pd.read_csv("Automobile.csv")
auto.head()
```

```python
sns.distplot(auto['city_mpg'],kde=False,rug=True);
```

```python
sns.jointplot(auto['horsepower'],auto['city_mpg']);
```

```python
sns.jointplot(auto['horsepower'],auto['city_mpg'],kind='hex');
```
