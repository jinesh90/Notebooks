```python
import pandas as pd
```

```python
# Formula for calculating number
"""
weighted rating
WR = (v/v+m)xR + (m/v+m)xC
where
v = # of votes 
m = is the minimum number of votes required for the movie to be in the chart (the prerequisite)
R = mean rating of movie
C is mean rating of all movies.
"""
```

```python
#Load movies into csv data frame
df_movie = pd.read_csv("/home/jinesh/Desktop/movies_metadata.csv")
```

```python
m = df_movie['vote_count'].quantile(0.80)
```

```python
# now filter movies who has runtime between 45 mins to 300 mins to qualify movies.

new_df_movies = df_movie[(df_movie['runtime'] >= 45) & (df_movie['runtime'] <= 300)]
```

```python
new_df_movies = new_df_movies[new_df_movies['vote_count'] > m]
#print(new_df_movies.shape)


# Calculate C
C = new_df_movies['vote_average'].mean()
```

```python
C
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    # Compute the weighted score
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
# Compute the score using the weighted_rating function defined above
new_df_movies['score'] = new_df_movies.apply(weighted_rating, axis=1)
```

```python
new_df_movies.sort_values(by=['score'], ascending=False)
```
