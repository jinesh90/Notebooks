```python
X,Y = mnist["data"],mnist["target"]
Y = Y.astype(np.uint8)
```

```python
X_train,X_test,Y_train,Y_test = X[:60000],X[60000:],Y[:60000],Y[60000:]
```

```python
Y_train_5 = (Y_train == 5)
```

```python
Y_test_5 = (Y_test == 5)
```

```python
from sklearn.linear_model import SGDClassifier
sdg_clf = SGDClassifier(random_state=42)
sdg_clf.fit(X_train,Y_train_5)
```

```python
for i in range(1500):
    print(i,sdg_clf.predict(X[i].reshape(1,-1)))
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(sdg_clf,X_train,Y_train_5,cv=3,scoring="accuracy")
```

```python
# Calculate Confusion matrix
from sklearn.model_selection import cross_val_predict
Y_train_predcit = cross_val_predict(sdg_clf,X_train,Y_train_5,cv=3)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_train_5,Y_train_predcit)
```

```python
y_scores = cross_val_predict(sdg_clf,X_train,Y_train_5, cv=3,method="decision_function")
y_scores
```

```python
from sklearn.metrics import precision_recall_curve
precison,recall,thresholds = precision_recall_curve(Y_train_5,y_scores)
```

```python
def plot_precision_recall_vs_thresold(precision,recall, thresold):
    plt.plot(thresold,precision[:-1],"b--",label="Precison")
    plt.plot(thresold,recall[:-1],"g-",label="Recall")
    plt.xlabel("Thresold")
    plt.legend(loc="center left")
    plt.ylim([0,1])
plot_precision_recall_vs_thresold(precison,recall,thresholds)
plt.show()
```

```python
from sklearn.metrics import precision_score,recall_score
Y_train_pred_90 = (y_scores < 500000)
precision_score(Y_train_5,Y_train_pred_90)
```

```python
recall_score(Y_train_5,Y_train_pred_90)
```
