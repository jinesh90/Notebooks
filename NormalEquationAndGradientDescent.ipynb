```python
# implement batch gradient descent
eta = 0.02 # Eta learning rate
n_iteration = 10000
n = 100 # no of features
theta_random = np.random.randn(2,1) # random intilization
print("Initial theta value is {}".format(theta_random))
```

```python
theta_random = np.random.randn(2,1) # random intilization
print("Initial theta value is {}".format(theta_random))
for i in range(n_iteration):
    gradient = 2/n * X_b.T.dot(X_b.dot(theta_random) - y)
    theta_random = theta_random - eta * gradient
    print ("After {} itration theta value is :{}".format(i+1,theta_random))
```

```python
# implement Stochastic gradient descent
n_iteration = 50
t0,t1 = 5,50 # learning schedule 
def learning_rate(t):
    return t0 / (t + t1)
```

```python
theta_stochastic = np.random.randn(2,1)
theta_stochastic
```

```python
for epoch in range(n_iteration):
    for i in range(5):
        random_index = np.random.randint(100)
        xi = X_b[random_index:random_index+1]
        yi = y[random_index:random_index+1]
        gradient = 2 * xi.T.dot(xi.dot(theta_stochastic) - yi)
        eta = learning_rate(epoch * 100 + i)
        theta_stochastic = theta_stochastic - eta * gradient
        print("After {} iteration theta is :{}".format(epoch, theta_stochastic))
```
