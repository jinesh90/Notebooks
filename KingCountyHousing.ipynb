```python
import pandas as pd
from sklearn.model_selection import train_test_split
housing = pd.read_csv("/home/jinesh/Desktop/kc_house_data.csv")
housing.head(10)
```

```python
housing.info()
```

```python
housing = housing.drop(["id","date","sqft_living15","sqft_lot15","yr_renovated"],axis=1)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
housing.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
train_set,test_set = train_test_split(housing,test_size=0.2,random_state=42)
train_set
```

```python
corr_matrix = housing.corr()
```

```python
corr_matrix['price'].sort_values(ascending=False)
```

```python
housing = train_set.copy()
```

```python
housing_price = train_set["price"].copy()
```

```python
housing = housing.drop(["price"],axis=1)
```

```python
housing.info()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
lin_reg = LinearRegression()
lin_reg.fit(housing,housing_price)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(lin_reg,housing,housing_price,scoring='neg_mean_squared_error',cv=100)
lin_rmse_score = np.sqrt(-scores)
lin_rmse_score.mean()
lin_rmse_score.std()
```

```python
lin_rmse_score.mean()
```

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor()
tree_reg.fit(housing,housing_price)
```

```python
scores = cross_val_score(tree_reg,housing,housing_price,scoring='neg_mean_squared_error',cv=100)
tree_rmse_score = np.sqrt(-scores)
tree_rmse_score.mean()
tree_rmse_score.std()
```

```python
tree_rmse_score.mean()
```

```python
# try another model called RandomForestRegressor which is building top on DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor()
forest_reg.fit(housing,housing_price)
```

```python
forest_score = cross_val_score(forest_reg,housing,housing_price,scoring="neg_mean_squared_error",cv=100)
forest_score_rmse = np.sqrt(-forest_score)
forest_score_rmse.mean()
```
